```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
incheon=pd.read_csv('incheon.csv')
kunsan=pd.read_csv('kunsan.csv')
```

```python
from plotnine import *
```

```python
incheon.year = incheon.year.astype('str')
```

```python
ggplot(incheon) + geom_line(aes(x='month',y='mean'),stat='identity') + facet_wrap('year') + scale_x_discrete(limits=list(range(1, 13)))
```

```python
ggplot(incheon) + geom_line(aes(x='month',y='mean',group = 'year',color='year'),stat='identity') 
```

```python
kunsan.year = kunsan.year.astype('str')
```

```python
ggplot(kunsan) + geom_line(aes(x='month',y='mean'),stat='identity') + facet_wrap('year') + scale_x_discrete(limits=list(range(1, 13)))
```

```python
ggplot(kunsan) + geom_line(aes(x='month',y='mean',group = 'year',color='year'),stat='identity') 
```

```python
incheon['city'] = 'incheon'
kunsan['city'] = 'kunsan'
```

```python
p = ggplot() + \
    geom_line(aes(x='month', y='mean', group='year', color = 'city'),data=incheon, stat='identity') + \
    geom_line(aes(x='month', y='mean', group='year', color = 'city'),data=kunsan, stat='identity') 

print(p)
```
