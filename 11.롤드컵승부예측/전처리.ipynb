### 10월 18일기준으로 20판 정도 게임이 진행되었으므로 밴픽합이 10이상이면 메타챔이라고 생각한다.

```python
hotchamp = banpick[banpick['밴+픽'] >= 10]
hotTeam = teamDF[teamDF.champ.isin(hotchamp.챔피언.unique())]
hotTeam = hotTeam.groupby(['team','champ']).agg({'pick':np.sum})
hotTeam.reset_index(inplace= True)
```

```python
hotDF = pd.DataFrame([])
for i in range(len(hotchamp.챔피언)):
    mask0 = hotTeam[hotTeam.champ == hotchamp.챔피언[i]].pick > hotchamp.픽[i]
    hotDF = pd.concat([hotDF, hotTeam[hotTeam.index.isin(mask0.index)]])
    hotDF.reset_index(drop=True,inplace=True)
```

```python
teamDF.dropna().groupby('team').agg({'pick':np.sum})
```

```python
hotDF = hotDF.groupby('team').agg({'pick':np.sum})
hotDF = pd.concat([hotDF,teamDF.dropna().groupby('team').agg({'pick':np.sum}) ],axis=1)
hotDF.columns = ['meta_pick','total_pick']
```
