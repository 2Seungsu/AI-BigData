```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df19 = pd.read_csv('players_19.csv')
df20 = pd.read_csv('players_20.csv')
df21 = pd.read_csv('players_21.csv')
```

```python
playerDF = pd.DataFrame([])
```

```python
df = pd.read_csv('player_final.csv')
```

```python
df.player_positions = df.player_positions.apply(lambda x:x.split(',')[0])
```

```python
df.player_positions[0].split(',')[0]
```

```python
df.player_positions[1].split(',')[0]
```

```python
playerDF['sofifa_id'] = df['sofifa_id']
playerDF['name'] = df['short_name']
playerDF['overall'] = df['overall']
playerDF['potential'] = df['potential']
playerDF['value_eur'] = df['value_eur']
playerDF['wage_eur'] = df['wage_eur']
playerDF['player_url'] = df['player_url']
playerDF['player_face_url'] = df['player_face_url']
playerDF['age'] = df['age']
playerDF['height_cm'] = df['height_cm']
playerDF['weight_kg'] = df['weight_kg']
playerDF['club'] = df['club_name']
playerDF['league'] = df['league_name']
playerDF['position'] = df['player_positions']
playerDF['attacking'] = df.loc[:,df.columns.str.contains('attacking')].mean(axis=1)
playerDF['skill'] = df.loc[:,df.columns.str.contains('skill')].mean(axis=1)
playerDF['movement'] = df.loc[:,df.columns.str.contains('movement')].mean(axis=1)
playerDF['power'] = df.loc[:,df.columns.str.contains('power')].mean(axis=1)
playerDF['mentality'] = df.loc[:,df.columns.str.contains('mentality')].mean(axis=1)
playerDF['defending'] = df.loc[:,df.columns.str.contains('defending')].mean(axis=1)
playerDF['goalkeeping'] = df.loc[:,df.columns.str.contains('goalkeeping')].iloc[:,:-1].mean(axis=1)
```

```python
playerDF.dropna(inplace = True)
```

```python
playerDF.to_csv('player_real_final.csv',index=False)
```
